## Dataset Overview

- The Dataset contains 183412 rows and 22 columns.
- There are 12 numerical data types, 2 datetime, 3 object types, 1 is a boolean type and 4 are category data types.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
GoBike = pd.read_csv('201902-fordgobike-tripdata-clean-version.csv')
```

## Trips during different times of day and different days of the week

> > The busiest day of the week is Thursday followed by Tuesday, Wednesday, Friday then Monday.

> Saturday and Sunday have the same amount of trips done.

>  Most trips were taken at 17 hours, followed by 08 hours.
The least trips were taken between 03 hours and 04 hours.
The distribution of the hours of day is binomial. Most trips takes place between 07 hours to 09 hours and from 16 hours to 18 hours. There aren't many trips from 24 hours to 05 hours.


```python
base_color = sb.color_palette()[0]
plt.figure(figsize=[16, 8])
```

```python
days = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
plt.figure(figsize=[8, 6])
sb.countplot(data = GoBike, x='day of the week', color = base_color, order = days);
plt.title("Number of trips each day");
```

```python
plt.figure(figsize=[8, 6])
sb.countplot(data = GoBike, x='time_of_day', color = base_color);
plt.xlabel('Hours');
plt.title("Count of trips in each hour");
```

##  Users' Gender, User_Type and Age

> The male gender was the most popular with significantly more than half the users of Ford GoBike compared to females and other.

> The majority of user types using Ford GoBike are subscribers

> Majority of users fall between the ages of 27 and 39



```python
plt.figure(figsize=[12, 7])

bins = np.arange(GoBike['age'].min(), GoBike['age'].max())

plt.hist(GoBike['age'], bins = bins);
plt.xlabel('Age');
plt.ylabel('Count');
plt.title("The age distribution of Ford GoBike users");
```

```python
genders = GoBike.member_gender.value_counts()

#Pie Distribution of the Gender
plt.figure(figsize=[12, 7])
plt.pie(genders, labels = genders.index, autopct='%1.1f%%')
plt.legend()
plt.title("Gender Distribution")
plt.axis('square');
```

```python
user_type = GoBike.user_type.value_counts()

#Pie Distribution of the User Type
plt.figure(figsize=[12, 7])
plt.pie(user_type, labels = user_type.index, autopct='%1.1f%%')
plt.title("User Type Distribution")
plt.axis('square');
```

## Relationship between User Type, Time of Day and Trip Duration

> Customers generally have a longer trip duration.
> Subscribers trip duration is more constant than Customers with an exception of the times between 12 AM and 5 AM



```python
#Point plot between time of day, duration_min and user_type
sb.pointplot(data=GoBike, x='time_of_day', y='duration_min', hue='user_type', linestyles="");
plt.xlabel('Time of Day');
plt.ylabel('Trip Duration in Minute');
```
